## Dataset Description

This dataset contains comprehensive information about smartphones. Each record represents a mobile phone along with its technical specifications and pricing details. The dataset consists of the following features:

- **Name** – Brand and model name of the smartphone.
- **Price** – Selling price of the smartphone.
- **Rating** – Overall performance rating of the smartphone.
- **SIM** – SIM type and supported network connectivity (2G, 3G, 4G, 5G, VoLTE, Wi-Fi, NFC, etc.).
- **Processor** – Processor name, number of cores, and clock speed.
- **RAM** – RAM capacity and internal storage.
- **Battery** – Battery capacity and fast charging support.
- **Display** – Screen size, resolution, and refresh rate.
- **Camera** – Rear and front camera specifications.
- **Card** – Memory card support and maximum expandable storage.
- **OS** – Operating system and Android version.
