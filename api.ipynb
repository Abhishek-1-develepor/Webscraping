```python
import pandas as pd
import requests
```

```python
def welcome():
   print("="*100)
   print("                              welcome to indian railway Enquiry                                       ")
   print("="*100)
   source = input("Enter Source Station Code: ")
   destination = input("Enter Destination Station Code: ")
   date = input("Enter Date (YYYY-MM-DD:")
   response = requests.get(
    f"https://api.railradar.in/v1/trains/between/{source}/{destination}?{date}",
    headers={"Authorization": "rg_cbe8f5cbecc04ddb9468597af1f11102"},
   )
   train_detail=response.json()
   print(pd.json_normalize(train_detail['data']['trains']))

```
